## Problema: Obter as 10 ações mais negociadas no Brasil

Fonte de dados: https://br.financas.yahoo.com/noticias/acoes-mais-negociadas

```python
import pandas as pd
pd.options.mode.chained_assignment = None # evitar alertas ao manipular o dataframe abaixo
```

O Pandas tem uma função que extrai diretamente as tabelas HTML de uma página para um dataframe.  
No caso, examinando o código-fonte da página verificamos que a primeira tabela é a que queremos.

```python
dfs = pd.read_html('https://br.financas.yahoo.com/noticias/acoes-mais-negociadas')
df = dfs[0]
df
```

A informação é ordenada pelo volume de negócios do dia, mas queremos o volume médio em 3 meses.

```python
volume = df[['Símbolo', 'Nome', 'Volume Médio (3 meses)']]
volume
```

Na tabela acima há um NaN no volume médio de 3 meses, significando provavelmente que este ativo é um `outlier` para nossa análise.  
Além disso, a informação na coluna 'Volume Médio (3 meses)' está como *string*, indicada em milhões. Como só precisamos da comparação, vamos apenas transformar em valor de ponto flutuante.

```python
volume = volume.dropna(subset = ['Volume Médio (3 meses)'])
volume['Volume Médio (3 meses)'] = volume['Volume Médio (3 meses)'].str.replace('M','')
volume['Volume Médio (3 meses)'] = volume['Volume Médio (3 meses)'].str.replace(',','.').astype(float)
volume
```

Ordenando pelo volume médio em 3 meses, do maior para o menor.

```python
volume.sort_values(by=['Volume Médio (3 meses)'], ascending=False, inplace=True)
volume
```

```python
volume.plot.bar('Símbolo', 'Volume Médio (3 meses)', figsize=(18,5));
```
