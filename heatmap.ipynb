### Mostrar apenas valores acima de um limite no *heatmap*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Criando um *dataframe* com valores aleatórios

```python
rng = np.random.default_rng()
uniform_data = pd.DataFrame(rng.standard_normal(size=(20, 20)), columns=list('ABCDEFGHIJLMNOPQRSTU'))
uniform_data
```

#### Criando os *labels* para o *heatmap*

```python
# Criando um novo dataframe com os valores que devem aparecer.
# No exemplo abaixo, tudo o que está além de 66% nos dois sentidos.
label_uniform_data = uniform_data[((uniform_data > 0.66) | (uniform_data < -0.66))]
# Numa tabela de correlação, a diagonal é preenchida por '1s', já que são a correlação dos dados com eles mesmos.
# Como não quero que apareçam, tiro todos os '1s' do dataframe
label_uniform_data = label_uniform_data[(label_uniform_data != 1)]
# Arrendondo os números para duas casa decimais
label_uniform_data = label_uniform_data.round(2)
# Transformando os valores númericos em strings para passar como labels
label_uniform_data = label_uniform_data.astype(str)
# Quando se 'dropa' os valores acima do dataframe, a célula é 'preenchida' com NaNs.
# Abaixo substituindo os NaNs por uma string vazia.
label_uniform_data = label_uniform_data.replace('nan','')
```

```python
# Acertando o tamanho do gráfico
f, ax = plt.subplots(figsize=(15, 10))
# Abaixo criando-se o 'heatmap' passando para 'annot' o dataframe com as strings que irão aparecer nas células.
# Importante: sem o campo "fmt = ''" o seaborn dá erro ao usar o dataframe de strings como label
# 'cmap' muda o esquema de cores.
ax = sns.heatmap(uniform_data, annot = label_uniform_data, fmt = '', cmap="binary")
```
