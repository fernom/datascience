### Mostrar apenas valores acima de um limite no *heatmap*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Criando um *dataframe* com valores aleat√≥rios

```python
rng = np.random.default_rng()
uniform_data = pd.DataFrame(rng.standard_normal(size=(20, 20)), columns=list('ABCDEFGHIJLMNOPQRSTU'))
uniform_data
```

#### Criando os *labels* para o *heatmap*

```python
# Criando um novo dataframe com os valores que devem aparecer.
# No exemplo abaixo, tudo o que est√° al√©m de 66% nos dois sentidos.
label_uniform_data = uniform_data[((uniform_data > 0.66) | (uniform_data < -0.66))]
# Numa tabela de correla√ß√£o, a diagonal √© preenchida por '1s', j√° que s√£o a correla√ß√£o dos dados com eles mesmos.
# Como n√£o quero que apare√ßam, tiro todos os '1s' do dataframe
label_uniform_data = label_uniform_data[(label_uniform_data != 1)]
# Arrendondo os n√∫meros para duas casa decimais
label_uniform_data = label_uniform_data.round(2)
# Transformando os valores n√∫mericos em strings para passar como labels
label_uniform_data = label_uniform_data.astype(str)
# Quando se 'dropa' os valores acima do dataframe, a c√©lula √© 'preenchida' com NaNs.
# Abaixo substituindo os NaNs por uma string vazia.
label_uniform_data = label_uniform_data.replace('nan','')
```

```python
# Acertando o tamanho do gr√°fico
f, ax = plt.subplots(figsize=(15, 10))
# Abaixo criando-se o 'heatmap' passando para 'annot' o dataframe com as strings que ir√£o aparecer nas c√©lulas.
# Importante: sem o campo "fmt = ''" o seaborn d√° erro ao usar o dataframe de strings como label
# 'cmap' muda o esquema de cores.
ax = sns.heatmap(uniform_data, annot = label_uniform_data, fmt = '', cmap="binary")
```
