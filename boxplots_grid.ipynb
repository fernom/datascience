### Plotar vários boxplots em linhas e colunas usando Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Criando um dataframe com dados aleatórios.

```python
rng = np.random.default_rng()
dados = pd.DataFrame(rng.standard_normal(size=(500, 15)), columns=list('ABCDEFGHIJLMNOP'))
dados
```

Laço para criar os *subplots* dentro da figura do Matplotlib.

```python
colunas = dados.columns.to_list()
plot_rows=3
plot_cols=5
fig, axs = plt.subplots(plot_rows, plot_cols,  figsize=(20, 15))
x = 0
for i in range(0, plot_rows):
    for j in range(0, plot_cols):
        atual = colunas[x]
        axs[i,j].boxplot(dados[atual])
        axs[i,j].set_xlabel(atual, rotation=0)
        axs[i,j].set(xticklabels=[])
        x=x+1
```
