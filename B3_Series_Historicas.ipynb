A *string* abaixo é um exemplo de um registro do arquivo de séries:

    012015042222MYPK11      010IOCHP-MAXIONBNS ORD NM   R$  000000000003000000000000300000000000030000000000003000000000000300000000000030000000000025000001000000000000015800000000000000474000000000000000009999123100000010000000000000BRMYPKN01OR1100

Este é um tipo de arquivo chamado de "comprimento fixo". O formato texto de comprimento fixo destina um número específico de caracteres por campo, não sendo utilizados separadores entre campos. Todos os registros tem o mesmo número de caracteres ainda que os valores dos registros caibam em menos caracteres.

Para ler este tipo de arquivo utiliza-se a função [pandas.read_fwf](https://pandas.pydata.org/docs/reference/api/pandas.read_fwf.html)  
Seguindo http://jonathansoma.com/lede/foundations-2017/pandas/opening-fixed-width-files/  
1) Criar uma lista com os comprimentos dos campos:

```python
comprimentos = [
2,  # TIPREG - TIPO DE REGISTRO - FIXO “01”
8,  # DATA-P - DATA DO PREGÃO - FORMATO “AAAAMMDD” 
2,  # CODBDI - CÓDIGO BDI - UTILIZADO PARA CLASSIFICAR OS PAPÉIS NA EMISSÃO DO BOLETIM DIÁRIO DE INFORMAÇÕES
12, # CODNEG - CÓDIGO DE NEGOCIAÇÃO DO PAPEL 
3,  # TPMERC - TIPO DE MERCADO - CÓD. DO MERCADO EM QUE O PAPEL ESTÁ CADASTRADO
12, # NOMRES - NOME RESUMIDO DA EMPRESA EMISSORA DO PAPEL 
10, # ESPECI - ESPECIFICAÇÃO DO PAPEL 
3,  # PRAZOT - PRAZO EM DIAS DO MERCADO A TERMO 
4,  # MODREF - MOEDA DE REFERÊNCIA - USADA NA DATA DO PREGÃO 
13, # PREABE - PREÇO DE ABERTURA PAPELMERCADO PREGÃO DO NO 
13, # PREMAX - PREÇO MÁXIMO PAPELMERCADO PREGÃO DO NO 
13, # PREMIN - PREÇO MÍNIMO DO PAPEL- MERCADO NO PREGÃO 
13, # PREMED - PREÇO MÉDIO DO PAPEL- MERCADO NO PREGÃO 
13, # PREULT - PREÇO DO ÚLTIMO NEGÓCIO DO PAPEL-MERCADO NO PREGÃO 
13, # PREOFC - PREÇO DA MELHOR OFERTA DE COMPRA DO PAPELMERCADO 
13, # PREOFV - PREÇO DA MELHOR OFERTA DE VENDA DO PAPELMERCADO 
5,  # TOTNEG - NÚMERO DE NEGÓCIOS EFETUADOS COM O PAPEL- MERCADO NO PREGÃO 
18, # QUATOT - QUANTIDADE TOTAL DE TÍTULOS NEGOCIADOS NESTE PAPEL- MERCADO 
18, # VOLTOT - VOLUME TOTAL DE TÍTULOS NEGOCIADOS NESTE PAPEL- MERCADO 
13, # PREEXE - PREÇO DE EXERCÍCIO PARA O MERCADO DE OPÇÕES OU VALOR DO CONTRATO PARA O MERCADO DE TERMO SECUNDÁRIO 
1,  # INDOPC - INDICADOR DE CORREÇÃO DE PREÇOS DE EXERCÍCIOS OU VALORES DE CONTRATO PARA OS MERCADOS DE OPÇÕES OU TERMO SECUNDÁRIO 
8,  # DATVEN - INDICADOR DE CORREÇÃO DE PREÇOS DE EXERCÍCIOS OU VALORES DE CONTRATO PARA OS MERCADOS DE OPÇÕES OU TERMO SECUNDÁRIO - FORMATO “AAAAMMDD”  
7,  # FATCOT - FATOR DE COTAÇÃO DO PAPEL - ‘1’ = COTAÇÃO UNITÁRIA ‘1000’ = COTAÇÃO POR LOTE DE MIL AÇÕES 
13, # PTOEXE - PREÇO DE EXERCÍCIO EM PONTOS PARA OPÇÕES REFERENCIADAS EM DÓLAR OU VALOR DE CONTRATO EM PONTOS PARA TERMO SECUNDÁRIO (ver planilha)
12, # CODISI - CÓDIGO DO PAPEL NO SISTEMA ISIN OU CÓDIGO INTERNO DO PAPEL (ver planilha) 
3,  # DISMES - NÚMERO DE DISTRIBUIÇÃO DO PAPEL (ver planilha) 
]
#print(comprimentos)
```

2) Criando uma lista com o nome das colunas.  
   A primeira linha do arquivo não é o nome das colunas (ver o documento da B3 referido acima) então precisamos nomear elas após a leitura do arquivo. Estou usando o código abreviado da referência.

```python
colunas = ['TIPREG', 'DATA-P', 'CODBDI', 'CODNEG', 'TPMERC', 'NOMRES', 'ESPECI', 'PRAZOT', 'MODREF', 'PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'TOTNEG', 'QUATOT', 'VOLTOT', 'PREEXE', 'INDOPC', 'DATVEN', 'FATCOT', 'PTOEXE', 'CODISI', 'DISMES']
print(colunas)
```

7a) A coluna DATVEN tem algumas datas com o valor `99991231` que seria a data 31/12/9999. Acredito que seja assim para indicar ativos sem data de vencimento. Só que o Python não consegue tratar esta data, ver abaixo.

```python
teste['DATVEN'].value_counts()
```

https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timestamp-limitations  
**Limites para timestamp:**  
Since pandas represents timestamps in nanosecond resolution, the time span that can be represented using a 64-bit integer is limited to approximately 584 years:  

    In [92]: pd.Timestamp.min
    Out[92]: Timestamp('1677-09-21 00:12:43.145225')

    In [93]: pd.Timestamp.max
    Out[93]: Timestamp('2262-04-11 23:47:16.854775807')

Aqui (https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries-oob) sugere maneiras de tratar "out-of-bound" mas não acho necessário neste caso.  
**Para manter o dado no mesmo formato, acredito que subsituir "99991231" por um valor próximo a "Timestamp.max" é suficiente, indicando a substituição**.  

Será preciso verificar durante a utilização dos valores. 

```python
teste.loc[(teste['DATVEN'] == 99991231),'DATVEN'] = 22611231
teste['DATVEN'].value_counts()
```

```python
teste['DATVEN'] = pd.to_datetime(teste['DATVEN'], format='%Y%m%d')
teste.info()
```
